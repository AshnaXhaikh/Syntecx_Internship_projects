```python
import pandas as pd
import numpy as np
import os

# Create output directory
os.makedirs('outputs', exist_ok=True)
os.makedirs('data', exist_ok=True)

print(f"Pandas version: {pd.__version__}")
print(f"NumPy version: {np.__version__}")
```

```python
# Create a realistic sales dataset
np.random.seed(42)

# Generate sales data
data = {
    'Order_ID': [f'ORD-{i:05d}' for i in range(1, 1001)],
    'Date': pd.date_range('2024-01-01', periods=1000, freq='D'),
    'Product': np.random.choice(['Laptop', 'Phone', 'Tablet', 'Headphones', 'Charger'], 1000),
    'Category': np.random.choice(['Electronics', 'Accessories', 'Gadgets'], 1000),
    'Quantity': np.random.randint(1, 10, 1000),
    'Unit_Price': np.random.uniform(10, 1000, 1000).round(2),
    'Customer_City': np.random.choice(['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix'], 1000),
    'Customer_Type': np.random.choice(['New', 'Regular', 'VIP'], 1000, p=[0.3, 0.5, 0.2])
}

# Add calculated column
df = pd.DataFrame(data)
df['Total_Revenue'] = (df['Quantity'] * df['Unit_Price']).round(2)
df['Discount'] = np.random.choice([0, 5, 10, 15, 20], 1000, p=[0.6, 0.2, 0.1, 0.07, 0.03])
df['Final_Amount'] = (df['Total_Revenue'] * (1 - df['Discount']/100)).round(2)

# Save datasets
df.to_csv('data/sample_sales.csv', index=False)
df.to_excel('data/sample_sales.xlsx', index=False)
print("✅ Sample datasets created and saved")
print(f"Dataset shape: {df.shape}")
print(f"Columns: {list(df.columns)}")
```

```python
print("=== SUMMARY STATISTICS ===\n")

# 1. Basic statistics for numeric columns
print("1. BASIC STATISTICS")
print(f"Mean of Total_Revenue: ${df['Total_Revenue'].mean():.2f}")
print(f"Median of Total_Revenue: ${df['Total_Revenue'].median():.2f}")
print(f"Minimum Total_Revenue: ${df['Total_Revenue'].min():.2f}")
print(f"Maximum Total_Revenue: ${df['Total_Revenue'].max():.2f}")
print(f"Count of orders: {df['Total_Revenue'].count()}")
print(f"Standard Deviation: ${df['Total_Revenue'].std():.2f}")
print(f"Variance: ${df['Total_Revenue'].var():.2f}")

# 2. Multiple statistics at once
print("\n2. ALL STATISTICS AT ONCE")
print(df['Quantity'].agg(['mean', 'median', 'min', 'max', 'count', 'std']))

# 3. Statistics by group
print("\n3. GROUPED STATISTICS")
print("Average Revenue by Product:")
print(df.groupby('Product')['Total_Revenue'].mean().round(2))

print("\n4. PIVOT TABLE - Average Revenue by Category and Customer Type")
pivot_table = pd.pivot_table(df,
                             values='Total_Revenue',
                             index='Category',
                             columns='Customer_Type',
                             aggfunc='mean')
print(pivot_table.round(2))

# 5. Multiple aggregations
print("\n5. MULTIPLE AGGREGATIONS")
print(df.groupby('Customer_City').agg({
    'Total_Revenue': ['mean', 'sum', 'count'],
    'Quantity': 'mean'
}).round(2))

# 6. Cumulative statistics
print("\n6. CUMULATIVE STATISTICS")
df_sorted = df.sort_values('Date')
df_sorted['Cumulative_Revenue'] = df_sorted['Total_Revenue'].cumsum()
print("Cumulative revenue over time (first 10 days):")
print(df_sorted[['Date', 'Total_Revenue', 'Cumulative_Revenue']].head(10))

# 7. Summary statistics function
def get_summary_stats(df, column):
    """Custom function to get comprehensive statistics"""
    stats = {
        'Count': df[column].count(),
        'Mean': df[column].mean(),
        'Median': df[column].median(),
        'Mode': df[column].mode().iloc[0] if len(df[column].mode()) > 0 else None,
        'Min': df[column].min(),
        'Max': df[column].max(),
        'Range': df[column].max() - df[column].min(),
        'Std': df[column].std(),
        'Variance': df[column].var(),
        '25%': df[column].quantile(0.25),
        '50%': df[column].quantile(0.50),
        '75%': df[column].quantile(0.75),
        'IQR': df[column].quantile(0.75) - df[column].quantile(0.25)
    }
    return pd.Series(stats)

print("\n7. CUSTOM SUMMARY STATISTICS FOR Total_Revenue:")
print(get_summary_stats(df, 'Total_Revenue'))
```

```python
def filter_and_save(df, conditions, columns=None, output_name="filtered_data"):
    """
    Flexible function to filter and save data

    Parameters:
    - df: DataFrame
    - conditions: dict of column: condition pairs
    - columns: list of columns to select (None for all)
    - output_name: name for output file
    """

    filtered = df.copy()

    # Apply filters
    for col, condition in conditions.items():
        if isinstance(condition, tuple):
            # Range filter (min, max)
            min_val, max_val = condition
            filtered = filtered[(filtered[col] >= min_val) & (filtered[col] <= max_val)]
        elif isinstance(condition, list):
            # List filter
            filtered = filtered[filtered[col].isin(condition)]
        else:
            # Single value filter
            filtered = filtered[filtered[col] == condition]

    # Select columns
    if columns:
        filtered = filtered[columns]

    # Save results
    csv_path = f'outputs/{output_name}.csv'
    excel_path = f'outputs/{output_name}.xlsx'

    filtered.to_csv(csv_path, index=False)
    filtered.to_excel(excel_path, index=False)

    print(f"✓ Filtered {len(filtered)} rows")
    print(f"✓ Saved to {csv_path}")
    print(f"✓ Saved to {excel_path}")

    return filtered

# Example usage
print("=== Interactive Filtering Example ===\n")

# Filter 1: Premium electronics in NY
premium_ny = filter_and_save(
    df,
    conditions={
        'Category': 'Electronics',
        'Customer_City': 'New York',
        'Total_Revenue': (4000, 10000)  # Range filter
    },
    columns=['Order_ID', 'Product', 'Total_Revenue', 'Customer_Type'],
    output_name="premium_ny_electronics"
)

# Filter 2: High discount orders
high_discount = filter_and_save(
    df,
    conditions={
        'Discount': [15, 20],  # List filter
        'Customer_Type': 'Regular'
    },
    columns=['Order_ID', 'Product', 'Discount', 'Total_Revenue'],
    output_name="high_discount_regular"
)

print("\n✅ Interactive filtering completed!")
```
