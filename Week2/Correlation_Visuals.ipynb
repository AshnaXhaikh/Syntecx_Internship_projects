### Project Overview

This project focuses on identifying and visualizing relationships between numerical variables within a dataset. By utilizing Pearson correlation and distribution plots, we can determine how features interact with one another, which is a critical step in Feature Selection for Machine Learning.

### Objectives

* Compute a correlation matrix for all numerical features.
* Visualize the matrix using a annotated **Heatmap** to identify strong dependencies.
* Generate **Pairwise Plots** to observe the distribution and scatter patterns between variables.

---

### Implementation Code

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('titanic')

#Filter for numerical data only
numeric_df = df.select_dtypes(include=['number'])

# Create the Correlation Matrix
corr_matrix = numeric_df.corr()

# Plotting the Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='RdYlGn', linewidths=0.5)
plt.title('Feature Correlation Heatmap')
plt.show()
```

```python

# Plotting Pairwise Relationships
sns.pairplot(numeric_df, diag_kind='kde')
plt.show()
```

### **Negative Correlations**

* **Pclass & Fare (-0.55):** Strongest negative link; higher class numbers (3rd class) paid significantly lower fares.
* **Pclass & Survived (-0.34):** Indicates that 3rd-class passengers had a lower survival rate than 1st-class passengers.
* **Age & Pclass (-0.37):** Suggests older passengers were more likely to be in 1st or 2nd class.

### **Positive Correlations**

* **SibSp & Parch (+0.41):** Strongest positive link; passengers traveling with siblings/spouses were likely also traveling with parents/children.
* **Fare & Survived (+0.26):** Higher fares correlate with a higher probability of survival.

**Conclusion:** Socio-economic status (Fare and Pclass) was the most influential numerical factor in survival outcomes.
